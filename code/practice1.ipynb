```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt     
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv('../data/Students Social Media Addiction.csv')
data["Avg_Daily_Usage_Hours_cat"] = pd.cut(data["Avg_Daily_Usage_Hours"], bins=[0, 3, 5, 7, np.inf], labels=[1, 2, 3, 4])
# data["Avg_Daily_Usage_Hours_cat"].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Stratified Split
stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=3)
for train_index, test_index in stratified_split.split(data, data["Avg_Daily_Usage_Hours_cat"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]

# Random Split
random_split = data["Avg_Daily_Usage_Hours_cat"].sample(frac=0.8, random_state=3)


random_set_ratio = random_split.value_counts() / len(random_split)
strat_set_ratio = strat_train_set["Avg_Daily_Usage_Hours_cat"].value_counts() / len(strat_train_set["Avg_Daily_Usage_Hours_cat"])


print(len(strat_train_set["Avg_Daily_Usage_Hours_cat"]))
print(len(data["Avg_Daily_Usage_Hours"]))

```
